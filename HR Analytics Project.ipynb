We have chescked most of our data, but we havent checked the categorical variables, which I will now define and proceed to study.

```python
target = 'Attrition'
cat_vars = [var for var in data.columns if data[var].dtype == 'O' and var != target]

# There are few other categorical features which are not by default
# We will analyze those variables also
xtra_vars = ['Education', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobLevel' ,'JobSatisfaction',
             'PerformanceRating', 'RelationshipSatisfaction', 'WorkLifeBalance', 'NumCompaniesWorked',
             'StockOptionLevel', 'PercentSalaryHike', 'TrainingTimesLastYear']


cat_vars = cat_vars + xtra_vars
```

These are the categorical variables we have

```python
cat_vars
```

Time to plot all the categorical variables as count plots with attirition as a hue to check their impact.

```python
def plot_cat(var, dataframe):
    plt.figure(figsize=(16, 4))
    sns.countplot(x=var, hue=target, data=dataframe)
    plt.show()
```

```python
for i in cat_vars:
    plot_cat(i, data)
```

```python
fqs = num_vars + cat_vars + [target] # re defining our data.
```

```python
df = data[fqs]
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder # Converting categorical variables into numeric data
```

```python
enc = OrdinalEncoder()  
oc_data = enc.fit_transform(df[['OverTime','Attrition']].values) # these  are the data we need to transform
```

```python
# Drop those mentioned columns and replace them with oc_data
ord_cols = ['OverTime','Attrition']
oc_df = pd.DataFrame(oc_data, columns=ord_cols)
df = df.drop(columns=ord_cols)
df = pd.concat([df, oc_df], axis=1) 
```

```python
ohc = OneHotEncoder(sparse=False, drop='first') # Now removing these columns as they dont have much impact

ohc_cols = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
ohc_data = ohc.fit_transform(df[ohc_cols].values)

ohc_df = pd.DataFrame(ohc_data, columns=ohc.get_feature_names())
df = df.drop(columns=ohc_cols, axis=1)
df = pd.concat([df, ohc_df], axis=1)
```

```python
df.dtypes # Checking data types, we can proceed with splitting the data
```

```python
X = df.drop('Attrition', axis=1)
y = df['Attrition'].values # Defining X and y variables
```

##### Now to correct impalance, we will oversample the yes attrition variable

```python
from imblearn.over_sampling import SMOTE

oversample = SMOTE()
Xo, yo = oversample.fit_resample(X, y) 
```

```python
sns.countplot(x=yo)
```

We have fixed the data imbalance

```python
from sklearn.model_selection import train_test_split # splitting the data
X_train, X_test, y_train, y_test = train_test_split(Xo, yo, test_size=0.2, random_state=41)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test) # Standardising the data
```

## Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtr = DecisionTreeClassifier(random_state=1)
dtr.fit(X_train, y_train)
y_pred = dtr.predict(X_test)
dtr.score(X_test, y_test)
```

### Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfr = RandomForestClassifier(random_state=1)
rfr.fit(X_train, y_train)
y_pred_rfr = rfr.predict(X_test)
rfr.score(X_test, y_test)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
llr = LogisticRegression()
llr.fit(X_train, y_train)
y_pred_rfr = llr.predict(X_test)
llr.score(X_test, y_test)
```

### K nearest neighbors classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors =15)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
knn.score(X_test, y_test)
```

### Gradient boosting algorith

```python
from sklearn.ensemble import GradientBoostingClassifier
gba = GradientBoostingClassifier(random_state=1)
gba.fit(X_train, y_train)
y_pred_gba = gba.predict(X_test)
gba.score(X_test,y_test)
```
